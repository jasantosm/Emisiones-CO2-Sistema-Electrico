```python
import pydataxm as api_xm
import pandas as pd
import datetime as dt
```

```python
consult = api_xm.ReadDB()
```

## Descarga de datos Emisiones CO2

```python
df_emisiones_co2 = consult.request_data("EmisionesCO2", 0, dt.date(2013, 10, 1),dt.date(2021, 1, 4))
df_emisiones_co2
```

```python
df_emisiones_co2.to_csv('data/emisionesCO2.csv')
```
